## Task 04 â€“ Sentiment Analysis of Social Media Data

**Objective:**  
To analyze and visualize sentiment patterns in social media data to understand public opinion.

**Dataset:**  
Social Media Sentiment Dataset

**Tools Used:**  
Python, Pandas, Matplotlib, Seaborn, TextBlob, Jupyter Notebook


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
```

```python
df = pd.read_csv(r"R:\PRODIGY_DS_04\twitter_training.csv",header=None)
df.columns = ['id', 'topic', 'sentiment', 'text']
df['text'] = df['text'].astype(str)
df.head()

```

```python
def get_sentiment(text):
    try:
        text = str(text).lower()
        analysis = TextBlob(text)
        polarity = analysis.sentiment.polarity
        
        if polarity >= 0.1:
            return 'Positive'
        elif polarity <= -0.1:
            return 'Negative'
        else:
            return 'Neutral'
    except:
        return 'Neutral'

```

```python
#df['Predicted_Sentiment'] = df['text'].apply(get_sentiment)
#df.head()
df['Predicted_Sentiment'] = df['text'].apply(get_sentiment)
df['Predicted_Sentiment'].value_counts()
df.head()
```

```python
colors = ['yellow', 'red', 'gray']

df['Predicted_Sentiment'].value_counts().plot(
    kind='bar',
    color=colors
)

plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
df['Predicted_Sentiment'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)
plt.title('Sentiment Share')
plt.ylabel('')
plt.show()
```

## Observations

- Most social media posts were classified as Neutral, indicating mixed or low emotional expression.
- Some posts showed Positive sentiment, reflecting favorable opinions.
- Negative sentiment was less frequent compared to Neutral and Positive.
- Sentiment analysis helps understand public opinion trends from text data.
